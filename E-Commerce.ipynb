# 2) Sales by catogery

```python
Sales_by_category=data.groupby("Category")["Sales"].sum().reset_index()
```

```python
Sales_by_category
```

```python
labels=Sales_by_category["Category"].values
size=Sales_by_category["Sales"].values
```

```python
explode=[0,0,0.1]
plt.pie(size,labels=labels,shadow=True,autopct='%1.1f%%',explode=explode)
plt.title("Sales Analysis By Category")
plt.legend(loc="upper left")
plt.show()
```

# 3)Sales by Sub-category

```python
sales_by_subcatogery=data.groupby("Sub-Category")["Sales"].sum().reset_index()
```

```python
sales_by_subcatogery
```

# 4)Monthly profit analysis

# 5)Profit by category

```python
labels=profit_by_category["Category"].values
size=profit_by_category["Profit"].values
plt.pie(size,labels=labels,shadow=True,autopct='%1.1f%%')
plt.title("Profit Analysis By Category")
plt.legend(loc="upper right")
wedgeprops={'edgecolor' : 'black' ,'linewidth':9,'linestyle':'--'}
plt.show()
```

# 5)Profit for sub_category

```python
profit_by_sub_category=data.groupby("Sub-Category")["Profit"].sum().reset_index()
profit_by_sub_category
```

```python
Sub_Category=profit_by_sub_category["Sub-Category"].values
profit=profit_by_sub_category["Profit"].values
plt.bar(Sub_Category,profit)
plt.xticks(rotation=65)
plt.xlabel("Sub-Category")
plt.ylabel("Profit")
plt.title("Profit Analysis by Sub Category")
```

# 6)Sales and Profit-Customer_segment

```python
sales_profit_by_segment=data.groupby("Segment").agg({"Sales":"sum" ,"Profit":"sum"}).reset_index()
sales_profit_by_segment
```

```python
import numpy as np
segment=sales_profit_by_segment["Segment"]
sales=sales_profit_by_segment["Sales"]
profit=sales_profit_by_segment["Profit"]
x=np.arange(len(segment))
width=0.45
plt.bar(x-width/2,sales,width=width,label="Sales")
plt.bar(x+width/2,profit,width=width,label="Profit")
plt.xlabel("Segment")
plt.ylabel("Amount")
plt.title("Sales and Profit by Segment")
plt.legend()
plt.xticks(x,segment)
plt.show()
```

# 7)Sales to Profit Ratio

```python
sales_profit_by_segment=data.groupby("Segment").agg({"Sales":"sum" ,"Profit":"sum"}).reset_index()
sales_profit_by_segment["sales_to_profit_ratio"]=sales_profit_by_segment["Sales"]/sales_profit_by_segment["Profit"]
print(sales_profit_by_segment[["Segment","sales_to_profit_ratio"]])
```
